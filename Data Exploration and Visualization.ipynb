## Creating Time Series with Line Charts

#### United States GDP Per Capita

```python
us = data[data.country == 'United States']
us
```

```python
# Create a Line Graph using plt 

plt.plot(us.year, us.gdpPerCapita) # plotting X=Year, Y=US GDP Per Capita
plt.title("GDP Per Capita Growth in the U.S") # Title your Graph
plt.ylabel('GDP Per Capita') # Label y axis
plt.xlabel('Year') # Label x axis
plt.show() # Show the graph
```

## Compare GDP Per Capita Growth in the U.S and China

```python
china = data[data.country == "China"]
china
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.plot(china.year, china.gdpPerCapita)
plt.title('GDP per capita')
plt.legend(['United States', 'China']) # Show legends 
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
# To compare GDP growth for both countries, 
# we need to calculate the growth rate, Growth Rate= GDP Per Capita for each year/ The first year available GDP Per Capita

# To get the first year GDP Per Capita
# iloc[n]= internger location 
us.gdpPerCapita.iloc[0]
```

```python
us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
china_growth = china.gdpPerCapita / china.gdpPerCapita.iloc[0] * 100
```

```python
plt.title('GDP Per Capita Growth (first year = 100)') # Title the Graph
plt.plot(us.year, us_growth) # Plotting US growth rate
plt.plot(china.year, china_growth) # Plotting China growth rate
plt.legend(['United State', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita growth')
plt.show()
```

## Export the Graph in Different Formats

```python
plt.title('GDP Per Capita Growth (first year = 100)')
plt.plot(us.year, us_growth)
plt.plot(china.year, china_growth)
plt.legend(['United State', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita growth')
# Export Graph in pdf format
plt.savefig('GDP Per Capita Growth in the U.S and China.pdf') # Note: it will be saved in the current folder
# Export Graph in png format
plt.savefig('GDP Per Capita Growth in the U.S and China.png')
```
